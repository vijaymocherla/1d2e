## Ploting the HF intracule

```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def read_matrix(filename): 
    file = open(filename,'r')
    n = int(file.readline())
    comment = file.readline()
    matrix = np.zeros((n,n))
    for k in range(n):
        line = file.readline().split()
        i,j = int(line[0])-1,int(line[1])-1
        matrix[i,j] = float(line[2])
    file.close()
    return matrix
```

```python
mo_file = '../z_6/alpha_1e00_beta_1e00/n_1024/scf_mo_coeff.txt'
mo_coeff = read_matrix(mo_file)
```

```python
mo_coeff.shape
```

```python
mo = mo_coeff.T[:,0]**2
nmo = len(mo)
density = np.kron(mo,mo).reshape(nmo,nmo)
```

```python
max = np.max(density)
plt.imshow(density, extent=[-15,15,-15,15], cmap='bwr_r', vmax=max, vmin=-max)
plt.colorbar()
```
