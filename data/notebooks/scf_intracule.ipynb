## Ploting the HF intracule

```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path='../z_1/alpha_1e-1_beta_1e-1/n_1024'
mo_file = os.path.join(path,'scf_mo_coeff.txt')
density_file = os.path.join(path,'scf_density.txt')
```

```python
def read_matrix(filename): 
    file = open(filename,'r')
    n = int(file.readline())
    comment = file.readline()
    matrix = np.zeros((n,n))
    for k in range(n):
        line = file.readline().split()
        i,j = int(line[0])-1,int(line[1])-1
        matrix[i,j] = float(line[2])
    file.close()
    return matrix
```

```python
density = read_matrix(density_file)
mo_coeff = read_matrix(mo_file)
```

```python
mo = mo_coeff.T[:,0]
nmo = len(mo)
```

```python
plt.imshow(np.kron(mo,mo).reshape(nmo,nmo), extent=[-15,15,-15,15], cmap='bwr_r')
plt.colorbar()
```
